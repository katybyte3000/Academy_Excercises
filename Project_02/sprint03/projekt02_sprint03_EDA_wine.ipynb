### Step 1: Weiß und Rot zusammen fügen - und neues Label Quality

```python
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])
```

```python
# create new column
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], categories=['low', 'medium', 'high'])
```

## R für Beziehung
Es ist wichtig zu wissen, in welcher Beziehung die Werte der x-Achse und die Werte der y-Achse stehen. Wenn keine Beziehung besteht, kann die lineare Regression nicht zur Vorhersage verwendet werden.

Diese Beziehung – der Korrelationskoeffizient – ​​wird genannt r.

```python
from scipy import stats

x = winelist['fixed_acidity']
y = winelist['quality']

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```
