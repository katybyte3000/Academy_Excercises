## EDA of Wine

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

red_wine = pd.read_csv("winequality-red.csv", sep =";")

#C:\Users\cumdi\Desktop\Bootcamp Visual Studio\Modul 1\Project_02\windedata\winequality-red.csv
# SEP-Argument seperiert die Daten, hier ist es  mit einem Semicolon

white_wine = pd.read_csv("winequality-white.csv", sep =";")
```

```python
red_wine.tail(10)
```

### Step 1: Weiß und Rot zusammen fügen - und neues Label Quality

```python
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])
```

```python
# create new column
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], categories=['low', 'medium', 'high'])
```

```python

white_wine.tail(5)
```

### Unterstrich einfügen - ACHTUNG-immer columns!!! angeben sonst geht es um das ganze Frame aber wir sprechen ja nur die erste Zeile an

```python
white_wine.columns = white_wine.columns.str.replace(' ', '_')
red_wine.columns = red_wine.columns.str.replace( ' ','_') #replace Leerzeichen mit Unterstrich
red_wine.head(10)
```

## Kategorie Weinart hinzufügen

```python
white_wine['winetype']= "white"
white_wine.head(10)
```

```python
red_wine['winetype']= "red"
red_wine.head()
```

## Eine Allwine Liste

```python

allwine = pd.concat([red_wine, white_wine], ignore_index=True)
allwine.head()
```

```python
allwine = allwine.drop_duplicates()
allwine.shape
```

```python
allwine.to_csv('winelistall.csv', index=False)
```
