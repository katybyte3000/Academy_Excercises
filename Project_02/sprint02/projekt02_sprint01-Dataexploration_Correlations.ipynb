### Beispiel für das auwählen nur bestimmter Spalten für die Heatmap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Beispiel DataFrame mit 12 numerischen Spalten
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [9, 8, 7, 6, 5],
    'E': [10, 11, 12, 13, 14],
    'F': [14, 13, 12, 11, 10],
    'G': [1, 2, 1, 2, 1],
    'H': [2, 3, 2, 3, 2],
    'I': [5, 6, 7, 8, 9],
    'J': [1, 3, 5, 7, 9],
    'K': [9, 7, 5, 3, 1],
    'L': [4, 3, 2, 1, 0]
}

df = pd.DataFrame(data)

# Wähle nur die Spalten aus, die du in der Heatmap sehen möchtest
selected_columns = ['A', 'B', 'C', 'D']  # Beispiel: Du möchtest nur Spalten A, B, C und D verwenden

# Berechne die Korrelation nur für diese ausgewählten Spalten
korrelation = df[selected_columns].corr()

# Erstelle die Heatmap
sns.heatmap(korrelation, annot=True, cmap='coolwarm')

# Zeige die Heatmap an
plt.show()
```
