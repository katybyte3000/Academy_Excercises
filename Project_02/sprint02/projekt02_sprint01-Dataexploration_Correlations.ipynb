### Beispiel für das auwählen nur bestimmter Spalten für die Heatmap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Beispiel DataFrame mit 12 numerischen Spalten
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [9, 8, 7, 6, 5],
    'E': [10, 11, 12, 13, 14],
    'F': [14, 13, 12, 11, 10],
    'G': [1, 2, 1, 2, 1],
    'H': [2, 3, 2, 3, 2],
    'I': [5, 6, 7, 8, 9],
    'J': [1, 3, 5, 7, 9],
    'K': [9, 7, 5, 3, 1],
    'L': [4, 3, 2, 1, 0]
}

df = pd.DataFrame(data)

# Wähle nur die Spalten aus, die du in der Heatmap sehen möchtest
selected_columns = ['A', 'B', 'C', 'D']  # Beispiel: Du möchtest nur Spalten A, B, C und D verwenden

# Berechne die Korrelation nur für diese ausgewählten Spalten
korrelation = df[selected_columns].corr()

# Erstelle die Heatmap
sns.heatmap(korrelation, annot=True, cmap='coolwarm')

# Zeige die Heatmap an
plt.show()
```

```python
heatwine = pd.DataFrame(red_wine)

# Berechne die Korrelation nur für diese ausgewählten Spalten
korrelation = heatwine.corr()

# Erstelle die Heatmap
fig, ax=plt.subplots(figsize=(20,20))
sns.heatmap(korrelation, cmap='PiYG', annot=True, vmax=1,vmin=-1, annot_kws={'fontsize':12})


# Zeige die Heatmap an

plt.show()
```

Starke Korrelation:

Nahe 1 oder -1: Eine Korrelation von +1 oder -1 bedeutet eine perfekte positive bzw. negative lineare Beziehung.
Werte zwischen 0,8 und 1 oder -0,8 und -1 gelten als starke Korrelation. In diesem Fall sollte man genauer hinschauen, da diese Variablen sehr stark miteinander verbunden sind.
Mittlere Korrelation:

Werte zwischen 0,5 und 0,8 oder -0,5 und -0,8 zeigen eine moderate Korrelation. Es könnte sich lohnen, die Ursachen oder Zusammenhänge zwischen den Variablen weiter zu untersuchen.

```python
heatwine = pd.DataFrame(red_wine)

# Berechne die Korrelation der numerischen Spalten
korrelation = heatwine.corr()

# Erstelle die obere Dreiecksmatrix (Maskierung)
matrix = np.triu(korrelation)

# Erstelle die Heatmap
fig, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(korrelation, cmap='PiYG', annot=True, vmax=1, vmin=-1, annot_kws={'fontsize': 9}, mask=matrix)

# Zeige die Heatmap an
plt.show()
```

### Anova

****Wenn die Ergebnisse Ihres statistischen Tests statistisch signifikant sind (basierend auf voreingestellten Signifikanzniveaus, z. B. wenn der erhaltene p-Wert unter dem Signifikanzniveau von 5 % liegt), können Sie die Nullhypothese zugunsten der Alternativhypothese ablehnen.****

```python
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])
```

```python
red_wine.tail()
```

```python
from scipy import stats

F, p = stats.f_oneway(
    red_wine[red_wine['quality_label'] == 'low']['alcohol'],
    red_wine[red_wine['quality_label'] == 'medium']['alcohol'],
    red_wine[red_wine['quality_label'] == 'high']['alcohol']
)
print('ANOVA test for mean alcohol levels across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```
