### PH Wert Vergleich von Rot und Wei√üwein

```python
#sns.boxplot(x = 'quality_label', y = 'pH', data = white_wine) 
sns.boxplot(x = 'quality_label', y = 'pH', data = allwine, hue = 'winetype', palette="rainbow") 
```

Wein weist einen pH-Wert zwischen 2,7 und 4,0 auf, wobei pH-Werte von 3,5 und h√∂her schon kritisch sind. Der ideale pH-Wert f√ºr mikrobiologisch stabilen Wein liegt zwischen 2,9 und 3,9.


Wei√üweine weisen in der Regel einen niedrigeren pH-Wert auf und liegen im sauren Bereich mit einem pH-Wert von etwa 3 bis 3,5. Die meisten Rotweine haben einen pH-Wert zwischen 3,5 und √ºber 4. Dieser Unterschied im pH-Wert tr√§gt dazu bei, dass Wei√üweine als frischer und lebendiger wahrgenommen werden, w√§hrend Rotweine eine etwas mildere S√§ure aufweisen.

Quelle:https://www.weinfreunde.de/magazin/weinwissen/saeuregehalt-im-wein-der-schluessel-zu-geschmack-und-qualitaet/#:~:text=Um%20den%20S%C3%A4uregehalt%20eines%20Weins,Skala%20von%200%20bis%2014.

### Der Boxplot zeigt das es im mittleren Qualit√§tsbereich Bereich einige Ausreisser gibt wo der PH Wert schon uber dem kritischen 3,5 liegt

Im Wein gilt: Je niedriger der pH-Wert, umso h√∂her die Weins√§ure, umso stabiler der Wein
Quelle:https://www.weinhalle.de/lexikon/ph-wert.html

## A View on fixed_acidity (Weins√§ure)

Die Messung der S√§ure im Wein in Gramm pro Liter
Eine weitere M√∂glichkeit, den S√§uregehalt eines Weins zu bewerten, besteht darin, die S√§ure in Gramm pro Liter auszudr√ºcken. Als Faustformel l√§sst sich gut merken, dass ein ***Wei√üwein, der mindestens sechs Gramm S√§ure pro Liter enth√§lt als frisch zu bezeichnen*** ist. Riesling oder Champagner kann aber auch locker acht oder sogar neun Gramm S√§ure pro Liter enthalten.
