```python
heatwine = pd.DataFrame(red_wine)


# Wähle nur die Spalten aus, die du in der Heatmap sehen möchtest
selected_columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar']  # Beispiel: Du möchtest nur Spalten A, B, C und D verwenden

# Berechne die Korrelation nur für diese ausgewählten Spalten
korrelation = heatwine[selected_columns].corr()

# Erstelle die Heatmap
sns.heatmap(korrelation, annot=True, cmap='Purples')

# Zeige die Heatmap an
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definiere die Bins
bins = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 7, 7.5, 8, 8.5]

# Erstelle eine Figur mit zwei Subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 Reihe, 2 Spalten

# Plot 1: Roter Wein
b = red_wine['residual_sugar']
axes[0].hist(b, bins=bins, color='red', edgecolor='black', linewidth=1.0)
axes[0].set_title('Sugar in Red Wine', fontsize=14)
axes[0].set_xlabel('Residual Sugar (g/L)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].tick_params(axis='both', labelsize=10)
axes[0].grid(True, linestyle='--', alpha=0.5)

# Plot 2: Weißer Wein
a = white_wine['residual_sugar']
axes[1].hist(a, bins=bins, color='yellow', edgecolor='black', linewidth=1.0)
axes[1].set_title('Sugar in White Wine', fontsize=14)
axes[1].set_xlabel('Residual Sugar (g/L)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].tick_params(axis='both', labelsize=10)
axes[1].grid(True, linestyle='--', alpha=0.5)

# Layout anpassen, um Überlappungen zu vermeiden
plt.tight_layout()

# Zeige die Plots
plt.show()
```
