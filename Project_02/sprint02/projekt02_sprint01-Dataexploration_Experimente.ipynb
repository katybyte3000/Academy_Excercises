### Let's now group our qualities into 3 categories: low, medium and high quality wines.

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

## PIE CHARTS -  Kreisdiagramme

Aufgabe 1: The first plot you should do is a pie chart, showing the quantity of red wines compared to the quantity of white wines.

```python
import matplotlib.pyplot as plt

# Create a dataframe with the data

data = {'Product': ['Red Wine', 'White Wine'], 'Anzahl': [len(red_wine["winetype"]), len(white_wine["winetype"])]}
winecolors= ['red', 'beige']
df = pd.DataFrame(data)

# Plot the pie chart
section_out = [0,0.2]
plt.pie(df['Anzahl'], labels=df['Product'], colors=winecolors, shadow=3, explode=section_out)

plt.show()
```

### Pie Chart

```python
# Falscher Ansatz

redcount = len(red_wine["winetype"])
print(redcount)

whitecount = len(white_wine["winetype"])
print(whitecount)

allcount = whitecount + redcount
print(allcount)

labels = ['Rotwein', 'Wei√üwein']
sizes = [redcount, whitecount]
colors = ['r', 'y']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

# Titel hinzuf√ºgen
plt.title('Vergleich der Anzahl der Zeilen in den DataFrames')

# Diagramm anzeigen
plt.show()

#L√∂sung ohne Rechnen

#yy = np.array([1599,4898])
#plt.pie(yy)
#plt.show()

#Probleme mit Lapels
```
